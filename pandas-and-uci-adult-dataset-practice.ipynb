```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('../input/adult.data.csv')
data.head()
```

```python
data.columns=data.columns.str.lower().str.replace('-','_')
```

```python
data.columns
```

**1. How many men and women (*sex* feature) are represented in this dataset?** 

```python
data['sex'].value_counts()
```

```python
data['sex'].value_counts(normalize=True)
```

**2. What is the average age (*age* feature) of women?**

```python
data[data['sex']=='Female']['age'].mean()
```

**3. What is the percentage of German citizens (*native-country* feature)?**

```python
data['native_country'].value_counts(normalize=True)*100
```

```python
(data['native_country']=='Germany').sum()/len(data)*100
```

**4-5. What are the mean and standard deviation of age for those who earn more than 50K per year (*salary* feature) and those who earn less than 50K per year? **

```python
data.groupby(['salary'])['age'].agg([np.mean,np.std])
```

**6. Is it true that people who earn more than 50K have at least high school education? (*education â€“ Bachelors, Prof-school, Assoc-acdm, Assoc-voc, Masters* or *Doctorate* feature)**

```python
pd.crosstab(data['salary'],data['education'])
```

**8. What is the proportion of those who earn a lot (>50K) greater: married or single men (*marital-status* feature)? Consider as married those who have a *marital-status* starting with *Married* (Married-civ-spouse, Married-spouse-absent or Married-AF-spouse), the rest are considered bachelors.**

```python
married_list=['Married-civ-spouse', 'Married-spouse-absent', 'Married-AF-spouse']
df_men_married_lot=data[(data['marital_status'].isin(married_list)) & (data['sex']=='Male')&(data['salary']=='>50K')]
```

```python
df_men_married_lot.head()
```

**9. What is the maximum number of hours a person works per week (*hours-per-week* feature)? How many people work such a number of hours, and what is the percentage of those who earn a lot (>50K) among them?**

```python
data['hours_per_week'].max()
```

```python
df_max_hour_per_week=data[data['hours_per_week']==data['hours_per_week'].max()]
```

```python
df_max_hour_per_week.groupby('salary')['hours_per_week'].count()/len(df_max_hour_per_week)
```

**10. Count the average time of work (*hours-per-week*) for those who earn a little and a lot (*salary*) for each country (*native-country*). What will these be for Japan?**

```python
data[data['native_country']=='Japan'].groupby(['native_country','salary'])['hours_per_week'].mean()
```
